```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['SimHei']  #解决中文显示乱码问题
plt.rcParams['axes.unicode_minus']=False

from tqdm import tqdm

#pd.options.display.max_columns = None
#pd.options.display.max_rows = None
```

```python
TextSet = []
with open("./PsyQA_full.txt", "r", encoding = 'utf-8') as f:  # 打开文件
    for line in f.readlines():
        tmp_list = line.split('",')
        tmp_list[-1] = tmp_list[-1].replace('\n','') #去掉换行符
        TextSet.append(tmp_list) #将每行数据插入data中
TextSet
```

```python
data1 = pd.DataFrame(TextSet)
len(data1)
```

```python
del data1[1]
```

分块处理标点

```python
for i in tqdm(range(0,200000)):
    if(data1.loc[i,0] == '['):
        data1 = data1.drop(labels=i)
    elif(data1.loc[i,0] == '{'):
        data1 = data1.drop(labels=i)
    elif(data1.loc[i,0] == '}'):
        data1 = data1.drop(labels=i)
    elif(data1.loc[i,0] == ']'):
        data1 = data1.drop(labels=i)
    pass
```

```python
for i in tqdm(range(0,len(data1))):
    if(data1.loc[i,0] == '],'):
        data1 = data1.drop(labels=i)
    elif(data1.loc[i,0] == '},'):
        data1 = data1.drop(labels=i)
    pass
data1 = data1.reset_index()
del data1['index']
```

因为时间太长 所以后面使用挂载的方法运行，故代码无运行记录

```python
data1['labels'], data1['content'] = data1[0].str.split('": ', 1).str
```

```python
del data1[0]
```

```python
data1['a']= data1['labels'].str.split('"', 1).str[1]
```

```python
data1['b'] = data1['content'].str.split('"', 1).str[1]
```

```python
del data1['labels']
del data1['content']
data1.columns = ['labels','content']
```

```python
for i in tqdm(range(0,len(data1))):
    if data1.loc[i,'labels'] != 'question' and data1.loc[i,'labels'] != 'description' and data1.loc[i,'labels'] != 'keywords' and data1.loc[i,'labels'] != 'answer_text':
        data1 = data1.drop(labels=i)
    pass
data1 = data1.reset_index()
del data1['index']
```

创建空列表，并向内填充处理好的文本，然后拼成dataframe

```python
question = []
description = []
keywords = []
answer_text = []
for i in tqdm(range(0,len(data1))):
    if data1.loc[i, 'labels'] == 'question':
        question.append(data1.loc[i, 'content'])
    if data1.loc[i, 'labels'] == 'description':
        description.append(data1.loc[i, 'content'])
    if data1.loc[i, 'labels'] == 'keywords':
        keywords.append(data1.loc[i, 'content'])
    if data1.loc[i,'labels'] == 'answer_text' and  data1.loc[i-1,'labels'] == 'keywords':
        answer_text.append(data1.loc[i, 'content'])
    pass
```

```python
data_temp1 = pd.DataFrame(question)
data_temp1.columns = ['question']
data_temp2 = pd.DataFrame(description)
data_temp2.columns = ['description']
data_temp3 = pd.DataFrame(keywords)
data_temp3.columns = ['keywords']
data_temp4 = pd.DataFrame(answer_text)
data_temp4.columns = ['answer_text']
```

```python
data_final = pd.merge(data_temp1,data_temp2,left_index=True,right_index=True)
```

```python
data_final = pd.merge(data_final,data_temp3,left_index=True,right_index=True)
data_final = pd.merge(data_final,data_temp4,left_index=True,right_index=True)
```

```python
outputpath='./PsyQA.csv'
data_final.to_csv(outputpath,sep=',', index=False, header=True, encoding='utf_8_sig')
```
